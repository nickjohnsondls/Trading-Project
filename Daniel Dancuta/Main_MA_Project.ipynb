```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import datetime as dt
```

```python
ticker1 = "TSLA"

start1 = dt.date(2010, 1, 1)
end1  = dt.date(2011, 1, 1)

ticker1_data = yf.download(ticker1, start = start1, end = end1)
ticker1_data
```

```python
# 

average = 0
MAC = 0
# Access the 'High' and 'Low' for each index (date) and perform an operation
for date, row in ticker1_data.iterrows():
    high = row['High']
    low = row['Low']
    # Here you can perform any mathematical operation you want with high and low
    # For example, finding the average (mean) price of the day
    mean_price = (high + low) / 2
    average = average + mean_price
num_columns = len(ticker1_data.columns)
MAC = average / num_columns 
print(MAC)
```
