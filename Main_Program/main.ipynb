```python
# # Download historical data for TSLA
ticker = "TSLA"
start_date = dt.datetime(2019, 1, 1)
end_date = dt.datetime(2024, 1, 1)
data = yf.download(ticker, start=start_date, end=end_date)
```

```python
# # Save data to CSV to use with PyAlgoTrade
csv_file = 'STOCK_data.csv'  # Adjust path as needed
data.to_csv(csv_file)

# the data from the dataframe should now go into the stock_data .csv file 
```

```python
#Second ticker for Pairs trading stock 
tickerA = "KO"
tickerB = "PEP"
start = dt.date(2019, 1, 1)
end = dt.date(2024, 1, 1)
yf.download(tickerA, start=start, end=end).to_csv(tickerA + "_data.csv")
yf.download(tickerB, start=start, end=end).to_csv(tickerB + "_data.csv")


csv_fileA = 'AAPL_data.csv'
csv_fileB = 'MSFT_data.csv'
```

```python
# Parameters for each strategy
initial_cash = 10000
fast_ma_period = 50
slow_ma_period = 200
rsi_period = 14
fastKPeriod = 14   
slowDPeriod = 3 
windowSize = 10 # high frequency should be low like 5,10 , 15 - longer term strategies should be 20, 50 , 100, 200 
num_std_dev = 2
```

```python
# #Run three seperate strategies 
run_strategy(SMAStrategy, [ticker], [csv_file], initial_cash, fast_ma_period, slow_ma_period, ticker)
run_strategy(RSIStrategy, [ticker], [csv_file], initial_cash, rsi_period, ticker)
run_strategy(StochStrategy, [ticker], [csv_file], initial_cash, fastKPeriod, slowDPeriod, ticker)

# Run the PairsTradingStrategy and store the strategy instance
run_strategy(PairsTradingStrategy, [tickerA, tickerB], [csv_fileA, csv_fileB], initial_cash, windowSize, num_std_dev)
```
